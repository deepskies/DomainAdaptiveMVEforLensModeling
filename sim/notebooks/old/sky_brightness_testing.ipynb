```python
def vis_random_dist(data):
    fig, axs = plt.subplots(2, 4, figsize=(10, 5))

    # Set the random seed for reproducibility
    np.random.seed(0)

    # Generate random i values between 1 and 100
    i_values = np.random.randint(1, 101, size=(4, 4))

    for i in range(2):
        for j in range(4):
            title = axs[i, j].set_title("i = " + str(i_values[i, j]))
            im = axs[i, j].imshow(data[i_values[i, j]][0])
            axoff = axs[i, j].axis('off')

    plt.tight_layout()
    plt.colorbar(im, ax=axs.ravel().tolist())
    return fig, axs
```

```python
sky1 = np.load('sky1/CONFIGURATION_1_images.npy')
sky100 = np.load('sky100/CONFIGURATION_1_images.npy')
sky10000 = np.load('sky10000/CONFIGURATION_1_images.npy')
```

```python
fig, _ = vis_random_dist(sky1)
fig.suptitle("Sky Brightness = 1", x = 0.4, fontsize = 18, y = 1)
```

```python
fig, _ = vis_random_dist(sky100)
fig.suptitle("Sky Brightness = 100", x = 0.4, fontsize = 18, y = 1)
```

```python
fig, _ = vis_random_dist(sky10000)
fig.suptitle("Sky Brightness = 10000", x = 0.4, fontsize = 18, y = 1)
```
